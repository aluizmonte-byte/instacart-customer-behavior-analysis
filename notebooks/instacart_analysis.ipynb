```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
orders = pd.read_csv('instacart_orders (2).csv', sep=';')
products = pd.read_csv('products (1).csv', sep=';')
```

```python
orders.head()
```

```python
orders_by_hour = orders['order_hour_of_day'].value_counts().sort_index()

plt.figure(figsize=(10,5))
orders_by_hour.plot(kind='bar')
plt.title('Número de pedidos por hora do dia')
plt.xlabel('Hora do dia')
plt.ylabel('Número de pedidos')
plt.xticks(rotation=0)
plt.show()
```

```python
import zipfile

with zipfile.ZipFile('order_products.csv.zip', 'r') as zip_ref:
    zip_ref.extractall()
```

```python
order_products = pd.read_csv('order_products.csv', sep=';')
```

```python
top_products = order_products['product_id'].value_counts().head(10).reset_index()
top_products.columns = ['product_id', 'count']

top_products = top_products.merge(products[['product_id', 'product_name']], on='product_id')

top_products
```
